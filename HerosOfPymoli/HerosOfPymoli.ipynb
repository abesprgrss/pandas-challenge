```python
import pandas as pd

rawdata="Resources/purchase_data.csv"


purchase_data = pd.read_csv(rawdata)
df=pd.DataFrame(purchase_data)
df.head()
```

```python
#Total Number of Players
players=len(df["SN"].unique())
print("Total Players")
print("--------------")
print(players)
```

```python
#Number of Unique Items
items=len(df["Item Name"].unique())

#Average Purchase Price
avgp=round(df["Price"].mean(),2)

#Total Number of Purchases
pchstot=len(df)

#Total Revenue
rev=df["Price"].sum()

#PAcols= ["Unique Items", "Average Price", "Total Number of Purchases", "Total Revenue"]
#PAlist=[items, avgp, pchstot, rev]
#PAdf=pd.DataFrame(PAlist, columns=[PAcols])
#PAdf

#lst = [items, avgp, pchstot, rev]

#df2 = pd.DataFrame(lst, columns =['Unique Items', 'Average Price', 'Total Number of Purchases','Total Revenue']) 
#print(df2) 

PAdf = pd.DataFrame({
    "Unique Items": [items],
    "Average Price": [avgp],
    "Total Number of Purchases": [pchstot],
    "Total Revenue": [rev]
})
PAdf
```
